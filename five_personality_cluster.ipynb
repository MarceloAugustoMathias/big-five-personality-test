```python
kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(data)
```

```python
predicoes = k_fit.labels_
data['Clusters'] = predicoes
```

```python
data.head()
```

```python
data['Clusters'].value_counts()
```

```python
data.groupby('Clusters').mean()
```

```python
col_list = list(data.columns)
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]
```

```python
data_soma = pd.DataFrame()
data_soma['extroversion'] = data[ext].sum(axis=1)/10
data_soma['neurotic'] = data[est].sum(axis=1)/10
data_soma['agreeable'] = data[agr].sum(axis=1)/10
data_soma['conscientious'] = data[csn].sum(axis=1)/10
data_soma['open'] = data[opn].sum(axis=1)/10
data_soma['clusters'] = predicoes
```

```python
data_clusters = data_soma.groupby('clusters').mean()
data_clusters
```

```python
from math import pi

categories = list(data_clusters)
N = len(categories)
list_cluster = list(range(N))
colors = ['b', 'green', 'red', 'purple', 'green']
angles = [ n / float(N)*2*pi for n in range(N)]
angles = angles + angles[:1]


for i, c in zip(list_cluster, colors):
    plt.figure(figsize = (10,10))
    #ax = plt.subplot(111,polar = True)
    ax = plt.subplot(2, 3, i+1, polar=True)
    ax.set_theta_offset(pi/2)
    ax.set_theta_direction(-1)
    plt.xticks(angles[:-1],categories)
    ax.set_rlabel_position(0)
    plt.yticks([0,2,4,6],["0","2","4","6"],color= "red", size = 7)
    plt.ylim(0,6)
    values = data_clusters.loc[i].values.flatten().tolist()
    values = values + values[:1]
    ax.plot(angles,values, linewidth = 1, linestyle="solid", label = "Cluster " + str(i))
    ax.fill(angles,values, c, alpha=0.1)
    plt.title = "Cluster " + str(i)
    plt.legend(loc = "upper left",bbox_to_anchor = (0.1,0.1))
    plt.show()
```
